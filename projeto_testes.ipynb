```python
import requests

# URL do arquivo que você deseja baixar
url = "https://download.inep.gov.br/educacao_superior/indicadores/resultados/2022/co"

# Nome do arquivo que será salvo no seu computador
nome_arquivo = "arquivo_baixado.ext"

# Envia uma requisição GET para baixar o arquivo
response = requests.get(url)

# Salva o conteúdo do arquivo localmente
with open(nome_arquivo, 'wb') as f:
    f.write(response.content)

print(f"Arquivo '{nome_arquivo}' baixado com sucesso!")
```

```python
import requests
from bs4 import BeautifulSoup
from re import sub
```

```python
url = "https://www.gov.br/inep/pt-br/acesso-a-informacao/dados-abertos/indicadores-educacionais/indicadores-de-qualidade-da-educacao-superior"

links = []
session = requests.get(url)
if not session.ok:
    raise FileNotFoundError(f"Error on read {url}")

soup = BeautifulSoup(session.content, "html.parser")

years_tabs = soup.find_all("div", class_ = "tab")
years = sorted([sub(r"[^0-9\-]", "", elem.text) for elem in years_tabs], reverse=True)
print(years)
for year in years[:4]:
    print(year)
    response = requests.get(f"{url}/{year}")
    print(response.status_code)
    if not response.ok:
            raise FileNotFoundError(f"Error on read INEP Enade {year}")
    soup = BeautifulSoup(response.content, "html.parser")

    link_all = soup.find_all("a", href=True)
    link = [link["href"] for link in link_all if "conceito_enade" in 
                    link["href"].lower() and "xlsx" in link["href"]][0]
    links.append(link)
```
